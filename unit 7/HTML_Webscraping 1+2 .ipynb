```python
#we need to get content for all 5 pages. so we need to run an iteration
# basically we want to create a function that goes and get content of 100 songs of each page. we inspected the page
# our function should have start, stop, and step arguments.
# start with 1 and scrape until 500 the end of page 5 and step every 100 

songs=[]
artists=[]
#pages = np.arange(1, 500, 100) #Return evenly spaced values within a given interval. this is very confusing let'S
#do range


for page in range (1,6): #we start with one so we do not get duplicates for first page
    url = "https://musicbrainz.org/series/b3484a66-a4de-444d-93d3-c99a73656905?page=" + str(page)
    response= requests.get(url)
    tree = etree.HTML(response.text)
    sleep(randint(2, 10))
    #print (page)
    song=tree.xpath('//*[@id="content"]/div[5]/table/tbody//tr/td[2]/descendant::a/bdi/text()')
    songs.extend(song)
    artist=tree.xpath('//*[@id="content"]/div[5]/table/tbody//tr/td[3]/descendant::a[1]/bdi/text()')
    artists.extend(artist)
   
      
```

## Wikipedia list

```python
# scraping all URL from wikipedia 
url2= 'https://en.wikipedia.org/wiki/Lists_of_songs'
```

```python
# now let's get content 
r2 = requests.get(url2)
r2.status_code # 200 status code means OK!
```

```python
# parse html (create the 'soup')
soup2 = BeautifulSoup(r2.content, "html.parser")
#check that the html code looks like it should
soup2
```

```python
#we are doing iteration dierctly to get a list of urls from the wikipedia page. 
#so what did we do? we did nested loops:
#1 created urls list, then went to div then ul and then a. so now we have list of urls 
urls = []
for div in soup2.find_all('div', attrs = {'div-col columns column-width'}):
    for ul in div.find_all('ul'):
        for li in ul.find_all('li'):
            for a in li.find_all('a'):
                urls.append(f"https://en.wikipedia.org{a['href']}")# f is a feature in python to put a variable in a string directly

print (urls)
```

```python
# go to all the other pages urls and scrape, and then we have to do the same as we did earlier for each new pagecontent
for url in urls:
    request= requests.get(url)
    pagecontent= BeautifulSoup(request.content, "html.parser")
    
print (urls[1])
    
```
